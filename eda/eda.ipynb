```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_json("data/spy_socio_data.json")

df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
plt.figure(figsize=(10,5))
df["source"].value_counts().plot(kind="bar")
plt.title("News Distribution by Source")
plt.xlabel("Source")
plt.ylabel("Number of Articles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Interpretation: Source-wise Distribution

The bar chart shows that local news platforms such as LiveNagpur and Sakal
publish a higher number of crime-related articles compared to national sources.
This indicates that local media plays a crucial role in reporting city-level
crime incidents.


```python
plt.figure(figsize=(8,5))
df["category"].value_counts().plot(kind="bar")
plt.title("Crime Category Distribution")
plt.xlabel("Crime Category")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

### Interpretation: Crime Category Analysis

Robbery and murder-related crimes are the most frequently reported categories.
This suggests that violent and property-related crimes dominate news coverage
in Nagpur, while narcotics and kidnapping cases are reported less frequently.


```python
plt.figure(figsize=(6,4))
df["credibility"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Credibility Distribution")
plt.ylabel("")
plt.show()
```

### Interpretation: Credibility Analysis

Most articles fall under medium to high credibility categories, indicating that
the dataset is largely reliable. This is due to the use of trusted news sources
and validation during scraping.


```python
df["scrapedAt"] = pd.to_datetime(df["scrapedAt"])
daily = df.groupby(df["scrapedAt"].dt.date).size()

plt.figure(figsize=(10,4))
daily.plot()
plt.title("Crime News Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Articles")
plt.tight_layout()
plt.show()
```

### Interpretation: Time-Series Trend

The time-series graph shows noticeable spikes in news reporting on certain dates.
These spikes usually correspond to major crime incidents or high-profile cases,
indicating increased media activity during such events.


```python
df["hasImage"] = df["media"].apply(lambda x: len(x) > 0)

df["hasImage"].value_counts().plot(kind="bar")
plt.title("Articles With / Without Images")
plt.show()
```

### Observations
- Robbery and murder are the most reported crimes
- Majority articles are medium to high credibility
- News spikes indicate major crime events
